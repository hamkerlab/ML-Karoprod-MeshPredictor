```python
%matplotlib inline
plt.rcParams['figure.dpi'] = 150

def visualize(x, y):
    plt.figure()
    plt.imshow(y[0, :, :].T)
    plt.title("Thickness")
    plt.colorbar()
    plt.figure()
    plt.imshow(y[1, :, :].T)
    plt.title("epseqpl")
    plt.colorbar()
    plt.figure()
    plt.imshow(y[2, :, :].T)
    plt.title("Thinning")
    plt.colorbar()
    plt.show()
```

```python
x, y = reg.predict({
        'Blechdicke': 1.01, 
        'Niederhalterkraft': 400.0, 
        'Ziehspalt': 2.4, 
        'Einlegeposition': -5, 
        'Ziehtiefe': 30,
        'Stempel_ID': 3,
        'E': 191.37245,
        'Rp0': 138.22696,
        'Rp50': 449.528189,
    }, 
    positions=(500, 500))

visualize(x, y)
```
